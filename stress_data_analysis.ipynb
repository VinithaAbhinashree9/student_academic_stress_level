```python
# Basic EDA and visualization for the academic stress dataset
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
```

```python

stress = pd.read_csv('academic Stress level - maintainance 1.csv')
```

```python
stress.head()
```

```python
stress.describe(include='all')
```

```python
# Select numeric columns for stress-related scores
numeric_cols = [
    'Peer pressure',
    'Academic pressure from your home',
    'What would you rate the academic  competition in your student life',
    'Rate your academic stress index '
]

# Descriptive statistics
desc_stats = stress[numeric_cols].describe()
desc_stats

# Correlation matrix
corr_matrix = stress[numeric_cols].corr()
corr_matrix

# Set style
sns.set(style='whitegrid')
```

```python
# 1. Histogram for stress index
plt.figure(figsize=(6,4))
sns.countplot(x='Rate your academic stress index ', data=stress, palette='viridis')
plt.title('Distribution of Academic Stress Index')
plt.xlabel('Stress Index (1=Low, 5=High)')
plt.ylabel('Number of Students')
plt.tight_layout()
plt.show()
```

```python
# 2. Boxplots for all numeric stress variables
plt.figure(figsize=(8,5))
sns.boxplot(data=stress[numeric_cols], palette='Set2')
plt.title('Boxplot of Stress-related Ratings')
plt.xticks(rotation=30, ha='right')
plt.tight_layout()
plt.show()
```

```python
# 3. Correlation heatmap
plt.figure(figsize=(10,8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Between Stress Factors')
plt.xticks(rotation=30, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# 4. Stress index by academic stage
plt.figure(figsize=(6,4))
sns.boxplot(x='Your Academic Stage', y='Rate your academic stress index ', data=stress, palette='pastel')
plt.title('Academic Stress Index by Academic Stage')
plt.xlabel('Academic Stage')
plt.ylabel('Stress Index')
plt.tight_layout()
plt.show()
```

```python
# 5. Stress index by study environment
plt.figure(figsize=(6,4))
sns.boxplot(x='Study Environment', y='Rate your academic stress index ', data=stress, palette='coolwarm')
plt.title('Academic Stress Index by Study Environment')
plt.xlabel('Study Environment')
plt.ylabel('Stress Index')
plt.tight_layout()
plt.show()
```
